## Importing Libraries

```python
#importing pandas & numpy libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

# Disable warnings
import warnings
warnings.filterwarnings('ignore')
```

## Importing processed Data

```python
df2 = pd.read_csv('processed_data.csv', na_values=['?',"n.a.","NA","n/a", "na", None])
df2.head()
```

```python
#dropping Unnamed column as we don't Need this 

df2.drop(['Unnamed: 0'], axis=1, inplace=True)
df2.head(5)
```

##  Identifying top 10 handsets used by the customers.


```python
# Count the frequency of each handset
handset_counts = df2['Handset Type'].value_counts()

# Get the top 10 handsets
top_10_handsets = handset_counts.nlargest(10)

print(top_10_handsets)
```

## We can Better understand with graph

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
top_10_handsets.plot(kind='bar',colormap='PuOr')
plt.title('Top 10 Handsets by Frequency')
plt.xlabel('Handset Type')
plt.ylabel('Frequency')
#plt.xticks(rotation=0)
plt.show()
```

##  Identifying the top 3 handset manufacturers


```python
# Count the frequency of each manufacturer
manufacturer_counts = df2['Handset Manufacturer'].value_counts()

# Get the top 3 manufacturers
top_3_manufacturers = manufacturer_counts.nlargest(3)

print(top_3_manufacturers)
```

```python
# Create a bar plot
plt.figure(figsize=(8,6))
top_3_manufacturers.plot(kind='bar',colormap='RdBu')
plt.title('Top 3 Handset Manufacturers')
plt.xlabel('Handset Manufacturers')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

##  Identifying the top 5 handsets per top 3 handset manufacturer


```python
#we have a dataframe df with columns 'Handset Manufacturer' and 'Handset Type'
top_manufacturers = df2['Handset Manufacturer'].value_counts().nlargest(3).index

top_handsets = df2[df2['Handset Manufacturer'].isin(top_manufacturers)].groupby('Handset Manufacturer')['Handset Type'].apply(lambda x: x.value_counts().nlargest(5)).reset_index()

print(top_handsets)
```

```python
# Create a list of manufacturers
manufacturers = ['Apple', 'Samsung', 'Huawei']

# Create a figure with multiple subplots
fig, axs = plt.subplots(len(manufacturers), 1, figsize=(10, 15))

# Loop through each manufacturer
for i, manufacturer in enumerate(manufacturers):
    # Filter the data for the current manufacturer
    df_manufacturer = df2.loc[df2['Handset Manufacturer'] == manufacturer]
    
    # Group by handset type and count the occurrences
    df_manufacturer = df_manufacturer.groupby(['Handset Type']).size().reset_index(name='count')
    
    # Get the top 5 handsets
    df_manufacturer = df_manufacturer.nlargest(5, 'count')
    
    # Plot the pie chart
    axs[i].pie(df_manufacturer['count'], labels=df_manufacturer['Handset Type'], autopct='%1.1f%%', startangle=90)
    axs[i].set_title(f"Top 5 {manufacturer} Handsets")
    
    # Equal aspect ratio ensures that pie is drawn as a circle
    axs[i].axis('equal')

# Layout so plots do not overlap
fig.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create a list of manufacturers
manufacturers = ['Apple', 'Samsung', 'Huawei']

# Loop through each manufacturer
for manufacturer in manufacturers:
    # Filter the data for the current manufacturer
    df_manufacturer = df2.loc[df2['Handset Manufacturer'] == manufacturer]
    
    # Group by handset type and count the occurrences
    df_manufacturer = df_manufacturer.groupby(['Handset Type']).size().reset_index(name='count')
    
    # Get the top 5 handsets
    df_manufacturer = df_manufacturer.nlargest(5, 'count')
    
    # Plot the bar chart
    ax.bar(df_manufacturer['Handset Type'], df_manufacturer['count'], label=manufacturer)

# Set the title and labels
ax.set_title("Top 5 Handsets by Manufacturer")
ax.set_xlabel("Handset Type")
ax.set_ylabel("Count")

# Rotate x-axis labels by 45 degrees
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Legend
ax.legend()

# Show the plot
plt.show()
```

```python
numerical_cols = df2.select_dtypes(include=[np.number])

# Calculate the IQR for each numerical column
Q1 = numerical_cols.quantile(0.25).values
Q3 = numerical_cols.quantile(0.75).values
IQR = Q3 - Q1

# Calculate the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = (numerical_cols < lower_bound) | (numerical_cols > upper_bound)

# Print the outliers
print(outliers.sum())
```
